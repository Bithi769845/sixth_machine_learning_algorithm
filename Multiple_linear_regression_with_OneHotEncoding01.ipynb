```python
df=pd.read_csv('/content/gdrive/MyDrive/Dataset/carprices.csv')
df.head()
```

```python
df.shape
df.isnull().sum()
```

```python
#One Hot Encoding
dummy_variables = pd.get_dummies(df['Car Model'],drop_first=True)
dummy_variables.head()
```

```python
new_df=df.drop('Car Model',axis=1)
new_df.head()
```

```python
df=pd.concat([new_df,dummy_variables],axis=1)
df.head()
```

```python
y=df['Sell Price']
y.head()
```

```python
x=df.drop('Sell Price', axis=1)
x.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(x,y)
```

```python
reg.score(x,y)
```

```python
reg.predict([[4500,8,0,0,1]])
```
